# Bar Graph

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(25,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df.plot(kind='bar')
```

```python
my_df.abs().plot(kind='bar')
```

```python
my_df.abs().plot(kind='bar',stacked=True,grid=True)
```

```python
my_df.abs().plot(kind='bar',stacked=True,grid=True,alpha=0.4)
```

```python
my_df.abs().plot(kind='bar',stacked=True,grid=True,title="Bar Chart")
```

```python
my_df.abs().plot(kind='bar',stacked=True,grid=True,title="Bar Chart",legend=False)
```

```python
my_df.abs().plot.bar()
```

```python
my_df.abs().plot.bar(stacked=True,alpha=0.4,title='Bar Chart Second type')
```
