# Density and KDE Plots

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(500,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df.plot(kind='kde')
```

```python
my_df.plot.kde()
```

```python
my_df.plot.density()
```

```python
my_df.plot.kde(alpha=0.4,lw=2,grid=True,title="KDE")
```
