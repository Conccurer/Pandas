# Box Plot

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(500,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df.plot(kind='box')
```

```python
my_df.plot(kind='box',figsize=(10,5))
```

```python
my_df.plot(kind='box',colormap='plasma')
```

```python
my_df.plot(kind='box',title='Box Plot')
```

```python
my_df.plot.box(title="Box plot")
```

```python
my_df.plot(kind='box',grid=True)
```
