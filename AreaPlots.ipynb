# Area Plots

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(25,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df.plot(kind='area',stacked=False)
```

```python
my_df.abs().plot(kind='area')
```

```python
my_df.abs().plot(kind='area',alpha=0.3)
```

```python
my_df.abs().plot(kind='area',alpha=0.3,title='Area Plot')
```

```python
my_df.abs().plot(kind='area',alpha=0.3,title='Area Plot',legend=False)
```

```python
my_df['One'].abs().plot(kind='area',alpha=0.3,title='Area Plot',legend=False)
```

```python
my_df[['One','Two']].abs().plot(kind='area',alpha=0.3,title='Area Plot',legend=False)
```

```python
my_df[['One','Two']].abs().plot(kind='area',alpha=0.3,title='Area Plot',legend=False,table=True)
```

```python
my_df.abs().plot(kind='area',alpha=0.3,title='Area Plot',legend=True,table=True)
```
