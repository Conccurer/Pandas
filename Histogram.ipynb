# Histogram

```python
import pandas as pd
import numpy as np
from numpy.random import rand
%matplotlib inline
```

```python
my_df = pd.DataFrame(rand(100,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df['Three'].hist()
```

```python
my_df['Three'].hist(bins=20)
```

```python
my_df['Three'].plot(kind='hist',bins=20)
```

```python
my_df['Three'].hist(bins=20,grid=False,legend=True)
```
