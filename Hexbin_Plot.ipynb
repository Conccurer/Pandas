# Haxbin Plot

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(500,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df.plot(kind='hexbin',x="One",y='Two')
```

```python
my_df.plot(kind='hexbin',x="One",y='Two',gridsize=20)
```

```python
my_df.plot(kind='hexbin',x="One",y='Two',gridsize=10,colormap='Wistia')
```

```python
my_df.plot.hexbin(x='One',y='Two',C='Three',gridsize=10)
```

```python
my_df.plot.hexbin(x='One',y='Two',C='Three',gridsize=10,title='Hexbin Plot')
```

```python
my_df.plot.hexbin(x='One',y='Two',C='Three',gridsize=10,title='Hexbin Plot',grid=True)
```
