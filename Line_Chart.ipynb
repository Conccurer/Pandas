# Line Chart

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(25,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df.plot(kind='line')
```

```python
my_df.abs().plot(kind='line')
```

```python
my_df.abs().plot(kind='line',lw=5)
```

```python
my_df['One'].abs().plot(kind='line')
```

```python
my_df[['One','Four']].abs().plot(kind='line',lw=3)
```

```python
my_df[['One','Four']].abs().plot(kind='line',lw=3,figsize=(10,5),title='Line Chart of "One" and "Four"')
```

```python
my_df[['One','Four']].abs().plot(kind='line',lw=3,alpha=0.4)
```

```python
my_df.plot.line()
```

```python
my_df.abs().plot.line()
```
