# Scatter Plot

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(500,4),columns=['One','Two','Three','Four'])
my_df
```

```python
my_df.plot(kind='scatter',x='One',y='Four',c='Two')
```

```python
my_df.plot(kind='scatter',x='One',y='Four',c='Two',cmap='plasma')
```

```python
my_df.plot(kind='scatter',x='One',y='Four',s=my_df['Two']*100)
```

```python
my_df.plot(kind='scatter',x='One',y='Four',s=my_df['Two']*100,alpha=0.4)
```

```python
my_df.plot(kind='scatter',x='One',y='Four',s=my_df['Two']*100,alpha=0.4,title="Scatter Plot")
```

```python
my_df.plot.scatter(x='One',y='Two')
```
